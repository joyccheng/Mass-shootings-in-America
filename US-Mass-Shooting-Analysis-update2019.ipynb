```python
# cleaning both datasets, 
# getting only the columns I want to work with, 
# getting rid of the ones I don't need for this analysis, and
# match column names to better merge datas later

df2.columns = map(str.title, df2.columns)
# change all column names to title format

df_cleaned = df.drop(['Data Source 1', 'Data Source 2', 'Data Source 3',
                      'Data Source 4', 'Data Source 5', 'Data Source 6',
                      'Data Source 7', 'Number of Victim Fatalities', 'Description'
                      ], axis=1)
# get rid of columns that I don't need for this analysis

df2_cleaned = df2.rename(columns={'Case':'Title',
                                  'Prior_Signs_Mental_Health_Issues':'History of Mental Illness - General',
                                  'Mental_Health_Details':'History of Mental Illness - Detailed',
                                  'Location.1':'Place Type',
                                  'Age_Of_Shooter':'Average Shooter Age',
                                  'Race':'Shooter Race',
                                  'Fatalities':'Total Number of Fatalities',
                                  'Injured':'Number of Victims Injured',
                                  'Total_Victims':'Total Number of Victims',
                                  'Gender':'Shooter Sex'}) 
# renaming column names to match original datasets' column names

df2_cleaned['State'] = df2['Location'].str.split(',').str.get(1)
# to values of state from Location

df_cleaned.columns
df2_cleaned.columns
```

```python
def school(place):
    '''
    Returns a new series containing yes and/or no.
    
    Returns value 'yes' if the element contains 'School', 
    and 'no' if the element does not contain the word.
    
    Arg: Series
    
    '''
    if 'School' in place.title():
        return 'Yes'
    else:
        return 'No'
        
df2_cleaned['School Related'] = df2_cleaned['Place Type'].apply(school)
# Store the new Series into a new column in df2_cleaned dataset. 

df2_cleaned.head(1)
```

```python
def school_title(title):
    '''
    Distributing elements to its preferred category. 
     
    Arg: Series
        
    Search through each element in the Series to find if it contains 
    'university', 'elementary', 'middle school' or 'high school', 
    and rename element to the assigned names.
    If no matches occured, rename that element to 'Others'. 
    
    Returns: a new Series containing the returned values.
    
    '''

    if 'university' in title.lower():
        return 'College'
    elif 'college' in title.lower():
        return 'College'
    
    elif 'elementary' in title.lower():
        return 'Primary School'
    
    if 'middle school' in title.lower():
        return 'Secondary School'
    elif 'high school' in title.lower():
        return 'Secondary School'
    
    else:
        return 'Other'
        
df2_cleaned['Place Type'] = df2_cleaned['Title'].apply(school_title)
# store new values to Place Type
df2_cleaned['Place Type'].value_counts()
```

```python
# merge two data together by columns that have the same names
# but also inlcude columns that do not have matching names
data = df_cleaned.merge(df2_cleaned, how='outer', 
                        left_on= ['Title','Date', 'Location', 'State', 'Total Number of Victims',
                          'Number of Victims Injured','Total Number of Fatalities','Average Shooter Age',
                          'History of Mental Illness - General','History of Mental Illness - Detailed', 
                          'Shooter Race', 'Shooter Sex','School Related', 'Place Type', 'Latitude', 'Longitude'], 
                        right_on=['Title','Date', 'Location', 'State', 'Total Number of Victims',
                          'Number of Victims Injured','Total Number of Fatalities','Average Shooter Age',
                          'History of Mental Illness - General','History of Mental Illness - Detailed', 
                          'Shooter Race', 'Shooter Sex','School Related', 'Place Type', 'Latitude', 'Longitude'])
data.shape
data.head(20)
```

# Graph 4: What level of school had the most shootings? 


```python
school = data[['Title','School Related','Place Type']] 
school = school[ school['School Related'].str.lower().str.contains('yes') ]
# retrieve values from the rows that are school related 

def grade(s):
    '''Unifing strings to title format, 
    and distributing each element to its generalized category.
     
    Arg: Series
        
    Search through each element in the Series to find if it contains 
    college, primary, or secondary, and rename element to the assigned names.
    If no matches occured, rename that element to 'Others'. 
    
    '''
    if 'college' in s.lower():
        return 'College'    
    elif 'primary' in s.lower():
        return 'Primary'
    elif 'secondary' in s.lower():
        return 'Secondary'   
    else:
        return 'Others'

school['Place Type'].apply(grade).value_counts().iplot(kind='bar', 
                                                       xTitle= 'Grade', 
                                                       yTitle='Number of Shootings', 
                                                       title='What Grade of School Had the Most Shooting?',
                                                       color= '#5a96e6')
```

# Graph 5: Deaths vs. Victims? 

```python
data['Year'] = data['Date'].dt.year # store the years in the new column 

year_fatal_vict = data[['Year','Total Number of Fatalities','Number of Victims Injured']]
year_fatal_vict = year_fatal_vict.groupby('Year').sum()
# grouby Year to get the sum of total deaths and victims injured by year
# ex: 2019, total death: 57, injured: 76

year_fatal_vict.index
```

```python
# create trace0 to store valuse for the first plot
trace0 = go.Bar(x= year_fatal_vict.index,
                y = year_fatal_vict['Total Number of Fatalities'], 
                name= 'Killed',
                marker = {'color': '#bd0f0f'})

# create trace1 to store valuse for the second plot
trace1 = go.Bar(x= year_fatal_vict.index,
                y = year_fatal_vict['Number of Victims Injured'], 
                name = 'Wounded',
                marker = {'color': 'salmon'})
# combine two plots and store it in a list variable
victim_data = [trace0,trace1]

# define the layout for the graph
layout = {'title': 'Total Number of Fatalities and Victims Each Year',
          'xaxis':{'title':'Year'},
          'yaxis':{'title':'Number Killed and Wounded'},
          'barmode':'stack'}

# pass in the variable to plot two graphs into one using the created layout 
iplot({'data':victim_data, 'layout':layout})

```

# Graph 6 (Map): Which State has the highest number of Deaths from Public Mass Shooting? 


```python
import geopandas
import shapefile as shp
import seaborn as sns
```

#### Arguments:
     dict (
     type = 'choropleth',
     locations = list of states,
     locationmode= 'USA-States',
     colorscale = 'Earth' #predefined string or custom your own
     text = list or array of text to display per point
     z = array of values on z axis (color of state)
     colorbar = ('title' : 'Colorbar Titile'}
     )
     

```python
import numpy as np

data_map = data[['State','Total Number of Fatalities']]
data_map['State'] = data_map['State'].str.strip() # get rid of white spaces to avoid duplicates

geocodes = pd.read_csv('C:\\Users\\admin\\Desktop\\data-analysis-with-pandas\\us-state-ansi-fips.csv')
fipcode= geocodes.merge(data_map, how='left', left_on='stname', right_on= 'State').fillna(0.0) # reaplce all nan to 0.0

        
fip_group_sum = fipcode.groupby(['fips','State','abbrev']).sum()
fip_group_sum = fip_group_sum.add_suffix('_By State').reset_index()
# get the sum of total death by 'fips','State','abbrev'

states = fip_group_sum['abbrev'].str.strip() # get values from 'abbrev' for mapping, and strip out white spaces
values = fip_group_sum['Total Number of Fatalities_By State'] # get the numbers of total death for each state


fip_group_sum['text'] = (fip_group_sum['State'].astype(str) + ',' + \
                         fip_group_sum['abbrev'].astype(str) )
# create a column that will be used later in the map to display information in text
# ex: State, Abbrev
    
```

```python
# create a variable to store settings to define the map

mapping = [dict(type = 'choropleth', autocolorscale=True,
                locations= states, z= values, locationmode='USA-states',
                text= fip_group_sum['text'], 
                colorbar= dict(title='Numbers of Death')
               )]

mapping
```

```python
# setting up layout for the map

layout = dict(title='US States Deaths from Mass Shooting',
              geo = dict(scope = 'usa', projection= dict(type='albers usa'),
                         showlakes= True, lakecolor= 'rgb(66,165,245)'))
layout
```

```python
# display map

map_fig = dict(data=mapping, layout=layout)
iplot(map_fig)
```
